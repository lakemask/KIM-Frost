```python
'''
This program is RDPS + ML(SVM,RF)
Make by J.-H. Kang (2025.03)
Result : KIM-Rigional -> Frost
'''
import os
import glob
import pandas as pd
import numpy as np
import sys
import datetime
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import scipy.stats as stats
from sklearn.metrics import confusion_matrix
import pickle
from imblearn.over_sampling import BorderlineSMOTE
```

```python
# 73 class ver.
# INPUT DATA part

input_path = 'C:/Users/LG/Desktop/frost_data/ver3/ML_input'
input_list = glob.glob(f'{input_path}/**.pkl', recursive=True)
input_iter = len(input_list)

# +all data -> one table

for i in range(0,input_iter):
    with open(f'{input_path}/train_T{i}.pkl','rb') as f:
        input_data = pickle.load(f)
    X_train = input_data[:,:-2]
    Y_train = input_data[:,-2]
    print(Y_train)
    # Train data OverSampling

    smote = BorderlineSMOTE()
    X_train_new, Y_train_new = smote.fit_resample(X_train, Y_train)
    nb_0 = pd.DataFrame(Y_train_new).value_counts().iloc[0]
    nb_1 = pd.DataFrame(Y_train_new).value_counts().iloc[1]
    print("Finish Oversampling, check to count: Frost 0({}), Frost 1({})".format(nb_0, nb_1))

    scaler1 = MinMaxScaler()
    scaler1.fit(X_train_new)
    X_train_new = scaler1.transform(X_train_new)
    # save dataset (pickle files)

    output_path = 'C:/Users/LG/Desktop/frost_data/ver3/SMOTE_BL/73class'

    with open(f'{output_path}/X_train_T{i}.pkl','wb') as f1:
       pickle.dump(X_train_new, f1)
    with open(f'{output_path}/Y_train_T{i}.pkl','wb') as f2:
       pickle.dump(Y_train_new, f2)
    with open(f'{output_path}/scaler_T{i}.pkl','wb') as f5:
       pickle.dump(scaler1, f5)
```
